## parse  useful features

```python
%%time

# list of useful nutrients
main_nutrients = ['Energy (kCal)','Carbohydrates','Protein','Fat']

# define empty df
ingredients_df = pd.DataFrame()

for row in full_df.itertuples():
    
    #  build a dictionary - it can easily added by pandas :), even with missing features
    partial_dic = {}
    
    # get only FRENCH ingredients 
    if 'fr' in row.data['attributes']['ingredients-translations'].keys():

        step0 = row.data['attributes']['ingredients-translations']['fr']

        # split ingredients by different separators...
        # these include also '(' , ')', '[', ']' , as indeed often there are hierarchical!! ingredists list
        # unfortunately commas can still give some issues, as in https://www.openfood.ch/en/products/9138
        step1 = re.split(r',|\.|;|:|\(|\)|\[|\]|\{|\}|-|\r\n',step0)

        # remove: strings like : '1,7%', '2.7%' , etc...
        # remove: leading/ending white spaces          
        # remove * = usually used to indicate organic ingredient!
        # transform all carachters in LOWER CASE         
        step2 = [re.sub(r'\d*,*\.*\d*\s*%|\**','',my_str).strip().lower() for my_str in step1]
         
        
        # remove empty strings and coutry names!
        partial_ingre_list = [my_str for my_str in step2 \
                              if my_str not in ['','suisse','france','italie','europe','espagne','paraguay','allemagne','turquie','pérou']]

        #         store ingredients list
        partial_dic['ingredients'] = partial_ingre_list
        
        partial_dic['barcode'] = row.data['attributes']['barcode']
        partial_dic['name'] = row.data['attributes']['name']

    #     partial_dic['origins'] = row.data['attributes']['origins']
    #     partial_dic['quantity'] = row.data['attributes']['quantity']
    #     partial_dic['unit'] = row.data['attributes']['unit']

        [partial_dic.update({i['name']: i['per-hundred']}) for i in row.data['attributes']['nutrients'] if i['name'] in main_nutrients]


    
    ingredients_df  = ingredients_df.append(partial_dic,ignore_index=True)    
```

```python
# remove 1st step df
del full_df
gc.collect()
```

convert nutrients to `float` 

```python
ingredients_df[['Protein','Fat','Carbohydrates','Energy (kCal)']] = \
ingredients_df[['Protein','Fat','Carbohydrates','Energy (kCal)']].apply(lambda x: pd.to_numeric(x, errors='ignore'))
```

remove items with `null` ingredients

```python
ingredients_df = ingredients_df[ingredients_df.ingredients.notnull()]
```

# Build the network

## definition

Mind that as _ingredients network_, we mean the graph whose:

- nodes are _ingredients_
- edges are _number of items that contains that ingredient_

We start first with a __simplified(\*)__ version of the network:    
there is a link between 2 ingredients _if there is at least one product_  that contains both of them

(\*) As every product is basically the _complete subgraph of all his ingredients_, 
we can build the whole graph, by adding  subgraph (already implemented. feature in  `networkx` ;) )
